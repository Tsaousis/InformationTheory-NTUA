image.png

```python
import numpy as np
from fractions import Fraction

def Entropy(probs):
  entropy = 0
  for i in range(len(probs)):
      if probs[i] != 0:
          entropy -= probs[i] * np.log2(probs[i])
  return round(entropy,3)

def KL_distance(Pprobs,Qprobs):
  distance = 0
  for i in range(len(Pprobs)):
      if Pprobs[i] != 0:
          distance += Pprobs[i] * np.log2(Pprobs[i] / Qprobs[i])
  return round(distance,3)
```

Βρίσκω την περιθώρια κατανομή του Χ, Pr{X=i} για i=1,2,3 και την εντροπία Η(Χ)απο τον τύπο

```python

prx1=0.2+0+0.15
prx2=0.2+0.05+0.2
prx3=0.15+0.01+0.04


print("H(X) =",Entropy([prx1,prx2,prx3]))
```

Ομοίως και για το Η(Υ)

```python
pry1=0.2+0.2+0.15
pry2=0+0.05+0.2
pry3=0.15+0.2+0.04

print("H(Y) =",Entropy([pry1,pry2,pry3]))
```

Χρησιμοποιώντας τον παρακάτω τύπο έχουμε

image.png

```python
print("H(X|Y) = H(X,Y) - H(Y) =", Entropy(Excs2_1) - Entropy([pry1,pry2,pry3]))
print("H(Y|X) = H(X,Y) - H(X) =", Entropy(Excs2_1) - Entropy([prx1,prx2,prx3]))
```

Ακολουθώντας την μεθοδολογία και τους τύπους απο τις διαφάνειες έχουμε

```python
## Για Υ=1
Hx1_y1=0.2/pry1
Hx2_y1=0.2/pry1
Hx3_y1=0.15/pry1

print("H(X|Y=1) =", Entropy([Hx1_y1,Hx2_y1,Hx3_y1]))

## Για Υ=2
Hx1_y2=0.00/pry2
Hx2_y2=0.05/pry2
Hx3_y2=0.01/pry2

print("H(X|Y=2) =", Entropy([Hx1_y2,Hx2_y2,Hx3_y2]))

## Για Υ=3

Hx1_y3=0.15/pry3
Hx2_y3=0.2/pry3
Hx3_y3=0.04/pry3

print("H(X|Y=3) =", Entropy([Hx1_y3,Hx2_y3,Hx3_y3]))



print("H(Y) − H(Y|X) =", Entropy([pry1,pry2,pry3])-(Entropy(Excs2_1) - Entropy([prx1,prx2,prx3])))


print("D(x||y) =",KL_distance([prx1,prx2,prx3],[pry1,pry2,pry3]))

print("D(y||x) =",KL_distance([pry1,pry2,pry3],[prx1,prx2,prx3]))


print("I(Y;X) =",Entropy([prx1,prx2,prx3]) -  (Entropy(Excs2_1) - Entropy([pry1,pry2,pry3])))
```

# Άσκηση 3η

```python
p=[0.1,0.2,0.3,0.4]

q=[0.2,0.3,0.4,0.1]

r=[0.4,0.1,0.2,0.3]
print("1)")
print("H(X)=",Entropy(p))
print("H(Y)=",Entropy(q))
print("H(Z)=",Entropy(r))
```

```python
print("D(p||q) =",KL_distance(p,q))
print("D(q||p) =",KL_distance(q,p))
print("D(p||r) =",KL_distance(p,q))
print("D(r||p) =",KL_distance(q,p))
```

Παρατηρούμε ότι για την εντροπία οι τιμές της τυχαίας μεταβλητής δεν έχουν σημασία, ενώ για την απόσταση KullbackLeibler έχουν.
